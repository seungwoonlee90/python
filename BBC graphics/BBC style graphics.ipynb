# Load all the libraries & data you need

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/kirenz/datasets/master/gapminder.csv')
```

#  Two functions

```python
def bbc_plot(title, suptitle, yticks, grid) :
    plt.title(title, fontsize=28, fontweight="bold", loc='left', pad=60)
    plt.suptitle(suptitle, x=0.125, y=0.98, horizontalalignment="left", fontsize=22)
    plt.xticks(fontsize=18, color = "#222222")
    plt.yticks(yticks, fontsize=18, color = "#222222")
    plt.axhline(y=0, color='#333333', linewidth=5)
    plt.ylim(0)
    plt.grid(axis='y', linewidth=2,  color = "#cbcbcb", which=grid)
    ax.yaxis.set_ticks_position('none') 
    ax.xaxis.set_ticks_position('none')
    ax.tick_params(pad=15)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.show()
```

```python
def bbc_ploth(title, suptitle, xticks, grid) :
    plt.title(title, fontsize=28, fontweight="bold", loc='left', pad=60)
    plt.suptitle(suptitle, x=0.125, y=0.98, horizontalalignment="left", fontsize=22)
    plt.xticks(xticks, fontsize=18, color = "#222222")
    plt.yticks(fontsize=18, color = "#222222")
    plt.axvline(x=0, color='#333333', linewidth=5)
    plt.grid(axis='y', linewidth=2,  color = "#cbcbcb", which=grid)
    ax.yaxis.set_ticks_position('none') 
    ax.xaxis.set_ticks_position('none')
    ax.tick_params(pad=15)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    plt.show()
```

# Make a line chart

```python
line_df = df[(df['country'] == 'China')]
```

```python
fig, ax = plt.subplots(figsize = (10,6))

ax.plot(line_df['year'], line_df['lifeExp'], color = "#1380A1", linewidth=3)

title = "Living longer"
suptitle = "Life expectancy in China 1952-2007"
yticks = [0, 20, 40, 60]
grid = "major"

bbc_plot(title, suptitle, yticks, grid)
```

# Make a multiple line chart

```python
multiline_df = df[(df['country'] == 'China') | (df['country'] == 'United States')]
```

```python
fig, ax = plt.subplots(figsize = (10,6))

ax.plot(multiline_df['year'].unique(), multiline_df[multiline_df['country'] == 'China']['lifeExp'], color = "#FAAB18", linewidth=3, label="China")
ax.plot(multiline_df['year'].unique(), multiline_df[multiline_df['country'] == 'United States']['lifeExp'], color = "#1380A1", linewidth=3, label="United States")

title = "Living longer"
suptitle = "Life expectancy in China and the US"
yticks = [0, 20, 40, 60, 80]
grid = "major"

bbc_plot(title, suptitle, yticks, grid)
```

# Make a bar chart

```python
bar_df = df[(df['year'] == 2007) & (df['continent'] == 'Africa')].sort_values(by='lifeExp', ignore_index=True, ascending=False).head(5)
```

```python
fig, ax = plt.subplots(figsize = (10,6))

plt.bar(bar_df['country'], bar_df['lifeExp'], color = "#1380A1")

title = "Reunion is highest"
suptitle = "Highest African life expectancy, 2007"
yticks = [0, 20, 40, 60, 80]
grid = "minor"

bbc_plot(title, suptitle, yticks, grid)
```

## Make changes to the axes

```python
fig, ax = plt.subplots(figsize = (10,6))

plt.barh(bar_df['country'], bar_df['lifeExp'], color = "#1380A1")

title = "Reunion is highest"
suptitle = "Highest African life expectancy, 2007"
xticks = [0, 20, 40, 60, 80]
grid = "minor"

bbc_ploth(title, suptitle, xticks, grid)
```

# Make a histogram

```python
hist_df = df[df['year'] == 2007]
```

```python
fig, ax = plt.subplots(figsize = (10,6))

plt.hist(hist_df['lifeExp'], color = "#1380A1", range=(40, 90), bins=10)

title = "How life expectancy varies"
suptitle = "Distribution of life expectancy in 2007"
yticks = [0, 10, 20, 30]
grid = "minor"

bbc_plot(title, suptitle, yticks, grid)
```
