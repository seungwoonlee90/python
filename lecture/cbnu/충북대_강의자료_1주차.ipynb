```python
df_2
```

```python
df_2.to_csv('df_2.csv', encoding='cp949')
```

```python
df_2.columns = ['시도', '개소수', '이용량']
```

```python
df_2.head()
```

```python
df_2['시도'].replace('서울특별시', '서울')
```

```python
df_2.sort_values(by='이용량',ascending=False)
```

```python
df_2.corr()
```

```python
df_2['이용량/개소수'] =  df_2['이용량'] / df_2['개소수']
```

```python
df_2.sort_values(by='이용량/개소수',ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='AppleGothic')
plt.rc('axes', unicode_minus=False)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data = df_2, x="시도", y ='이용량')
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data = df_2.sort_values(by='이용량',ascending=False), x="시도", y ='이용량')
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data = df_2.sort_values(by='이용량',ascending=False), x="시도", y ='이용량',
            palette="Set1")
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data = df_2.sort_values(by='이용량',ascending=False), x="시도", y ='이용량',
            palette="Set4")
```

```python
plt.figure(figsize=(20,10))
plt.title('시도별 이용량 순위', fontsize=25)
sns.barplot(data = df_2.sort_values(by='이용량',ascending=False), x="시도", y ='이용량')
```

```python
plt.figure(figsize=(20,10))
plt.title('시도별 이용량 순위', fontsize=25)
plt.xlabel('우리나라 시도', fontsize=20)
plt.ylabel('이용량', fontsize=20)
sns.barplot(data = df_2.sort_values(by='이용량',ascending=False), x="시도", y ='이용량')
```
